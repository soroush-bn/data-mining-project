Previously we cleaned our data now it's time to tokenize each string and remove stop words such as the,a,...

```python

courses.index.stop
```

```python
courses["removed_stopwords"] = courses["useful"]
```

```python
courses.head(3)
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
for i in range(courses.index.stop):
  tokens = word_tokenize(courses.iloc[i,27])
  result = [x for x in tokens if not x in stop_words]
  courses.iloc[i,27]=str(result)

courses.head(3)
```

```python
courses.iloc[1,27]
```

so far we ebmedded 3 different columns, cleaned them with removing white spaces and punctuations and stop words and tokenized them using nltk module. Now we step further and we use stemming. 


Stemming is a process of reducing words to their word stem, base or root form (for example, books — book, looked — look). The main two algorithms are Porter stemming algorithm (removes common morphological and inflexional endings from words [14]) and Lancaster stemming algorithm (a more aggressive stemming algorithm)

```python
from ast import literal_eval
courses['removed_stopwords'] = courses['removed_stopwords'].apply(literal_eval)
courses.iloc[2,27]
```

```python
type(courses.iloc[2,27])
```

```python
courses["stemmed_removedStopWords"] = courses["removed_stopwords"]
```

```python
from nltk.stem import PorterStemmer
stemmer= PorterStemmer()
for i in range(courses.index.stop):
  result=[]
  for token in courses.iloc[i,28]:
    result.append(stemmer.stem(token))
  courses.iloc[i,28]=str(result)

courses.iloc[1,28]
    
```

**Lemmatization**

The aim of lemmatization, like stemming, is to reduce inflectional forms to a common base form. As opposed to stemming, lemmatization does not simply chop off inflections. Instead it uses lexical knowledge bases to get the correct base forms of words. 



```python
courses['stemmed_removedStopWords'] = courses['stemmed_removedStopWords'].apply(literal_eval)
courses.iloc[2,28]
```

```python
courses["Lemmanizated_removedStopWords"] = courses["removed_stopwords"]
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer=WordNetLemmatizer()
for i in range(courses.index.stop):
  result=[]
  for token in courses.iloc[i,27]:
    result.append(lemmatizer.lemmatize(token))
  courses.iloc[i,29]=str(result)

courses.iloc[2,29]
```

```python
courses['Lemmanizated_removedStopWords'] = courses['Lemmanizated_removedStopWords'].apply(literal_eval)
courses.iloc[2,29]
```

```python
courses["stemmedLem"] = courses["Lemmanizated_removedStopWords"]
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer=WordNetLemmatizer()
for i in range(courses.index.stop):
  result=[]
  for token in courses.iloc[i,30]:
    result.append(lemmatizer.lemmatize(token))
  courses.iloc[i,30]=str(result)

courses.iloc[2,30]
```

```python
import sklearn
import pandas as pd



lemSentences = courses['Lemmanizated_removedStopWords']
stemSentences = courses['stemmed_removedStopWords']
stemLemSentences = courses['stemmedLem']
lemSentences = lemSentences.apply(lambda x: ' '.join(x))
stemSentences = stemSentences.apply(lambda x:' '.join(x))
stemLemSentences = stemLemSentences.apply(lambda x:' '.join(x))
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.cluster import AgglomerativeClustering

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
model2 = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L6-v2')
lemembeddings = model.encode(lemSentences)
```

```python
print(lemembeddings)
```

```python
stemembeddings = model.encode(stemSentences)
```

```python
stemLemEmbeddings = model.encode(stemLemSentences)
```

# Clustering

```python
lemKmeans = KMeans(7,random_state=0).fit(lemembeddings)
lemDbscan = DBSCAN(eps=0.5,min_samples=3,metric="euclidean").fit(lemembeddings)
lemClusterWard = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="ward").fit(lemembeddings)
lemClusterSingle = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="single").fit(lemembeddings)
lemClusterComplete = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="complete").fit(lemembeddings)
```

```python
stemKmeans = KMeans(7,random_state=0).fit(stemembeddings)
stemDbscan = DBSCAN(eps=0.5,min_samples=3,metric="euclidean").fit(stemembeddings)
stemClusterWard = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="ward").fit(stemembeddings)
stemClusterSingle = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="single").fit(stemembeddings)
stemClusterComplete = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="complete").fit(stemembeddings)
```

```python
stemlemKmeans = KMeans(7,random_state=0).fit(stemLemEmbeddings)
stemlemDbscan = DBSCAN(eps=0.5,min_samples=3,metric="euclidean").fit(stemLemEmbeddings)
stemlemClusterWard = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="ward").fit(stemLemEmbeddings)
stemlemClusterSingle = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="single").fit(stemLemEmbeddings)
stemlemClusterComplete = AgglomerativeClustering(n_clusters=7,affinity="euclidean",linkage="complete").fit(stemLemEmbeddings)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import rand_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

labels = LabelEncoder().fit_transform(courses['Faculty'])
```
